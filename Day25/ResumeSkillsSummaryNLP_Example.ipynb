```python
#Step 1
#Import all required libraries

#To read csv file
import pandas as pd

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

import matplotlib.pyplot as plt
import seaborn as sns

#For training model
from sklearn.model_selection import train_test_split

#vectorizer
from numpy import vectorize
```

```python
df=pd.read_csv("/content/sample_data/UpdatedResumeDataSet.csv", encoding='utf-8')
print ("Shape of Data Frame: ",df.shape)
df.head(5)
```

```python
print ('\n Columns', df.columns.tolist())
```

```python
 sample_df=df.sample(10,random_state=42).reset_index(drop=True)
 #randomly select 10 rows from data frame
 sample_df
```

```python
sample_df=sample_df.rename(columns={'Resume':'Resume_Text'})
#Rename column from resume to resume_text

sample_df
```

```python
sample_df.shape
```

```python
vectorizer=TfidfVectorizer(stop_words='english',max_features=5000)
tfidf_matrix=vectorizer.fit_transform(sample_df['Resume_Text'])
tfidf_matrix.shape
```

```python
similaritiness_matrix=cosine_similarity(tfidf_matrix)
sim_df=pd.DataFrame(similaritiness_matrix, index=sample_df['Category']+"_"+sample_df.index.astype(str),
                    columns=sample_df["Category"]+"_"+sample_df.index.astype(str))
sim_df.round(2).head(3) #2 decimal places, first 3 rows
```

```python
similaritiness_matrix=cosine_similarity(tfidf_matrix)
sim_df=pd.DataFrame(similaritiness_matrix, index=df['candidate'],columns=df["candidate"])
sim_df.round(2)
```
