1. Spilts text into words/tokens (show the individual word token)
2. Clean Tokens (show clean token)
3. Remove Stop words (Show token after removing stop words)
4. Stemming (Show the words in their root forms)
5. Lemmatization (Show lemma words)


```python
import nltk, re
import pandas as pd
import numpy as np
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer

nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('wordnet') #needed for lemmatizer
```

```python
input_text="Natural Language Processing is going to help you out in near future . You have 2 remember this!! for getting interview calls"
```

```python
tokens=word_tokenize(input_text)
tokens
```

```python
clean_tokens=[re.sub(r'[^a-zA-z0-9\s]','',token.lower()) for token in tokens if token.isalpha()]
#^ meaning everything but, a-zA-z0-9]s - lower and uppercase alphabest, numbers, and whitespace, '' replace with nothing
clean_tokens
```

```python
stop_words=set(stopwords.words('english'))
filtered_tokens=[token for token in clean_tokens if token not in stop_words]
print ("Tokens after removing stop words:")
filtered_tokens
```

```python
stemmer = PorterStemmer()
stemmed_tokens=[stemmer.stem(token) for token in filtered_tokens]
stemmed_tokens
```

```python
lemmatizer = WordNetLemmatizer()
lemmatized_tokens=[lemmatizer.lemmatize(token) for token in filtered_tokens]
lemmatized_tokens
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer()
matrix=tfidf.fit_transform(lemmatized_tokens)
#matrix
#matrix.toarray()
df=pd.DataFrame(matrix.toarray(),columns=tfidf.get_feature_names_out())
df
```

TF-IDF Example 2

```python
corpus = ["Python developer required with strong coding skills",
          "Experienc in REST API, Automation, Jenkins, Flast and Python",
          "Looking for machine learning engineer familiar with NLP, Supervised, Unsupervised and Deep Learning"
          ]
```

```python
tfdif = TfidfVectorizer()
matrix =tfdif.fit_transform(corpus)
df=pd.DataFrame(matrix.toarray(),columns=tfdif.get_feature_names_out())
df
```

```python
print("\n Top Weighted Terms")
print(df.mean().sort_values(ascending=False).head(10))
```
