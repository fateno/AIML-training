# KNN Algorithm in Supervised Machine Learning 
    ## K-Nearest Neighbor(KNN) Algorithm
    ## KNeighborsClassifier of sklearn library

```python
import numpy as np 
from sklearn.neighbors import KNeighborsClassifier
```

```python
#data collection and splitting the data
X = np.array([
    [1,3500],
    [2,3200],
    [3,3000],
    [4,2500],
    [5,2300],
    [6,2000],
    [7,2200],
    [8,3700],
])
y=['Unfit','Unfit', 'Unfit','Fit','Fit','Fit','Fit','Unfit']
```

```python
#Create the model 
model=KNeighborsClassifier(n_neighbors=3)
```

```python
model.fit(X,y)
```

```python
new_person=[[4.5,7600]]
pred = model.predict(new_person)
print ("Predicted Result:", pred[0])
```

```python
exercise_hours=float(input("Enter number of exercise hours: "))
cal_intake=float(input("Enter number of calories consumed: "))
pred_fit_unfit=model.predict([[exercise_hours,cal_intake]])
print(f'Exercise Hours {exercise_hours}, Calories Intake {cal_intake}, You are likely {pred_fit_unfit[0]}')
```

```python
import pandas as pd 
```

```python
df = pd.read_csv("fruit_data.csv")
```

```python
X = df[['Weight','Size']]
y = df['Fruit']
```

```python
model=KNeighborsClassifier(n_neighbors=3)
model.fit(X,y)
```

```python
#new_fruit = [[120,2]]
#Another way to override the warning 
new_fruit=pd.DataFrame([[120,2]], columns=['Weight', 'Size'])
pred = model.predict(new_fruit)
print ("Predicted result: ", pred[0])
```

```python
fruit_weight=float(input("enter weight in gm"))
fruit_size=float(input("enter size in cm"))
our_fruit=pd.DataFrame([[fruit_weight, fruit_size]], columns=['Weight','Size'])
pred_fruit=model.predict(our_fruit) 
print (pred_fruit[0])
```

```python
X = df[['Hours_Studied','Assignments_Completed']]
y = df['Grade']
```

```python
model=KNeighborsClassifier(n_neighbors=3)
model.fit(X,y)
```

```python
new_student=pd.DataFrame([[1.5,4]], columns=['Hours_Studied','Assignments_Completed'])
pred=model.predict(new_student) 
print ("Predicted Result: ",pred[0])
```
