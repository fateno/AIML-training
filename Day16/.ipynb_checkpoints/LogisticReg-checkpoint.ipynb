```python
import pandas as pd 
import numpy as np 
from sklearn.linear_model import LogisticRegression
```

```python
#collect the data
df = pd.read_csv("pass_fail.csv")
df.head()
```

```python
#spill the data
X = df[['Hours']]
y = df[['Result']]
```

```python
#create the model (change here only)
model=LogisticRegression()
```

```python
model.fit(X,y)
```

```python
pred_35=model.predict([[3.5]])
print ("Predicted result for 3.5 hours:", pred_35[0])
```

```python
hours_study=int(input('Number of hours studied: '))
pred_value=model.predict([[hours_study]])
print (f"If you study for {hours_study}, you will {pred_value[0]}")
```

```python
import pandas as pd 
import numpy as np 
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

df=pd.read_csv("customer_purchase.csv")
df.head()
```

```python
le = LabelEncoder()
df['Purchased'] = le.fit_transform(df['Purchased'])
X = df[['Age', 'Salary']]
y = df['Purchased']
```

```python
model=LogisticRegression()
```

```python
model.fit(X,y)
```

```python
new_customer = [[28, 35000]]
prediction = model.predict(new_customer)
print (f"Predicted Result (1=Yes, 0=No): {prediction[0]}")
```

```python
new_cust_age=int(input('Enter Age'))
new_cust_sal=int(input('Enter Salary'))
new_cust=[[new_cust_age, new_cust_sal]]
prediction=model.predict(new_cust)
if (prediction):
    print ('Customer Purchased')
else: 
    print ('Customer Not Purchased')
```

```python
import pandas as pd 
import numpy as np 
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.feature_extraction.text import CountVectorizer

cv=CountVectorizer()
df=pd.read_csv("spam_dataset.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
le = LabelEncoder()
df['is_spam'] = le.fit_transform(df['is_spam'])
X = df[['message_content']]
x_data=cv.fit_transform(X)
y = df[['is_spam']]
```

```python
model=LogisticRegression()
```

```python
model.fit(x_data,y)
```
