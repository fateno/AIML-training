```python
import numpy as np 
from sklearn.tree import DecisionTreeClassifier,plot_tree
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_promo = { 
    "Experience_Years": [2,4,5,3,6,5],
    "Performance_Rating":[3,4,5,3,4,4],
    "Promoted":["No","Yes","Yes","No","Yes",None]
}
df_promo=pd.DataFrame(data_promo)
print("Dataset:")
df_promo
#display(df_promo)
```

```python
#clean the data 
train_df=df_promo.dropna()
```

```python
train_df
```

```python
#split data find input and output
X = train_df[['Experience_Years','Performance_Rating']]
y = train_df['Promoted']
```

```python
#Create the model 
model=DecisionTreeClassifier()
```

```python
#train the model 
model.fit(X,y)
```

```python
promotion=pd.DataFrame([[5,4]], columns=['Experience_Years','Performance_Rating'])
pred_promote=model.predict(promotion)
print("Promotion prediction: ",pred_promote[0])
```

```python
plt.figure(figsize=(6,4))
plot_tree(model, 
          feature_names=['Experience_Years', 'Performance_Rating'],
          class_names=['No', 'Yes'],
          filled=True)
plt.title('Decision Tree - Employment Promotion')
plt.show()
```

```python
emp_year=float(input("please enter your years of experience: "))
emp_rating=float(input("please enter your rating: "))
emp_promote=pd.DataFrame([[emp_year,emp_rating]], columns=['Experience_Years','Performance_Rating'])
pred_promote=model.predict(emp_promote)
#print(pred_promote[0])
#print (f"{emp_year} years of experience + {emp_rating} rating = {pred_promote[0]}")
if (pred_promote[0]) == "Yes": 
    #print ("Yay, promotion")
    print (f"{emp_year} years of experience + {emp_rating} rating = You're promoted")
else: 
     print (f"{emp_year} years of experience + {emp_rating} rating = You're not promoted")
```
