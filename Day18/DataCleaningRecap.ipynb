```python
#Step 1 - download dataset and print
import pandas as pd
```

```python
df = pd.read_csv("students_performance_dirty.csv")
df.head()
```

```python
#Get basic information
print ("---Dataset Information--- \n")
print (df.info())
```

```python
#Step3 - Print missing values per column
print ("---Missing values per column--- \n")
print (df.isnull().sum())
```

```python
#Step 4 - Check missing data line by line 

print ("Method 1")
for index,row in df.iterrows():
    if row.isnull().any(): 
        print(f'Row {index}: Missing Data -- {row.isnull().to_dict()} \n')

print ("Method 2")
missing_rows = df[df.isnull().any(axis=1)]
print("Rows with missing data:")
print(missing_rows)
```

```python
# Step 5 - Drop missing rows 
df_clean = df.dropna()
```

```python
# Step 6 - Compare before and after
print("Before dropping missing rows:", df.shape)
print("After dropping missing rows:", df_clean.shape)
```

```python
# Step 7 - Fill missing values 
df['filled_study_hours'] = df['study_hours'].fillna(df['study_hours'].mean())
df['gender_filled'] = df['gender'].fillna("unknown")

df_filled=df.fillna({
    'gender':'Unknown',
    'study_hours': df['study_hours'].mean()
})
```

```python
# Step 7 - Compare before and after 
#print("Before dropping missing rows:", df['study_hours'])
#print("After dropping missing rows:", df['filled_study_hours'])

print ("--First method of comparison--")

original = df['study_hours']
filled = df['filled_study_hours']

comparison = pd.DataFrame({
    'Original': original,
    'Filled': filled
})
    
#print(comparison.head(10))

nan_filled_hr = comparison[comparison['Original'].isna()]

print(nan_filled_hr)
print("*****")

originalg = df['gender']
filledg = df['gender_filled']

comparison_g = pd.DataFrame({
    'Original_g': originalg,
    'Filled_g': filledg
})

nan_filled_g = comparison_g[comparison_g['Original_g'].isna()]

print(nan_filled_g)
```

```python
print ("--Second method for comparison (Mr. Salman's example--")
for i in range(len(df)):
    print (f"\nRow (i) Before: {df.iloc[i].to_dict()}")
    print (f"Row (i) After: {df_filled.iloc[i].to_dict()}")
```
