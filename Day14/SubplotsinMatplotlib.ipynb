# Matplotlib.pyplot.subplots()
    ##subplots() - provide a way to plot multiple plots on single figure
    ## Give number of rows and columns -- Returns a tuple of (fig,ax) giving a single figure fig with an array of axes ax. 

```python
import matplotlib.pyplot as plt 
```

```python
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2, ncols=2, figsize=(10,5))
```

```python
import numpy as np 
sales_data={'Electronics':45000, 'Clothing':30000, 'Grocery':25000, 'Mobile':20000}
category=list(sales_data.keys())
actual_sales=list(sales_data.values())
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2, ncols=2, figsize=(10,7))
x=np.random.random(10)
#Define each plot
ax1.set_title('ax1: using random x value')
ax1.plot(x,x/2)
ax2.set_title('ax2: scatter plot using random x value')
ax2.scatter(x,x)
ax3.set_title('ax3: bar chart using sales data')
ax3.bar(category,actual_sales)
ax4.set_title('ax4: histograph random normal distribution value')
ax4.hist(np.random.randn(1000))
```

# Subplots using indexing

```python
our_lines=np.linspace(0,10,100)
#linspace - generates array of evenly spaced numbers over specified intervals
#(start,end,number of samples-default 50)
```

```python
product_prices={'Hair oil':20, 'Face Wash':30, 'Lip Balm':10, 'Face Cream':15}
products_list= list(product_prices.keys())
prices_list= list(product_prices.values())
```

```python
our_numbers=np.random.randn(100)
```

```python
our_dots=np.random.randn(1000)
```

```python
fig,ax = plt.subplots(nrows=2, ncols=2, figsize=(8,5))
ax[0,0].plot(our_lines, our_lines/2)
ax[0,1].bar(products_list, prices_list,color=['lavender', 'pink', 'salmon', 'skyblue'])
ax[1,0].scatter(np.random.randn(100), np.random.randn(100), color="orange")
ax[1,1].hist(our_dots,color="red")
plt.tight_layout()
```

```python
fig,ax = plt.subplots(nrows=2, ncols=2, figsize=(8,5))
ax[0,0].plot(our_lines, our_lines/2)
ax[0,1].bar(products_list, prices_list,color=['lavender', 'pink', 'salmon', 'skyblue'])
ax[1,0].scatter(our_numbers, our_numbers, color="orange")
ax[1,1].hist(our_dots,color="red")
plt.tight_layout()
```
