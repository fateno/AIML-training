```python
import pandas as pd
import matplotlib.pyplot as plt 
df= pd.read_csv('netflix_data.csv')
df=df.dropna(subset=['release_year'])
df['release_year']=df['release_year'].astype(int)
plt.figure(figsize=(10,5))
plt.hist(df['release_year'],bins=30, color='lightgreen', edgecolor='pink', alpha=0.8)
plt.title('Netflix released titles by year')
plt.xlabel('Release year')
plt.ylabel('Number of titles')
plt.grid(color='yellow', linestyle='--', linewidth=0.8)
plt.tight_layout()
plt.savefig('netflix_bar.pdf', dpi=300, bbox_inches='tight')
plt.show()
```

```python
df=df.dropna(subset=['release_year', 'type'])
df['release_year']=df['release_year'].astype(int)
movies=df[df['type']=='Movie']
print(movies['show_id'].count())
shows=df[df['type']=='TV Show']


plt.figure(figsize=(10,5))
plt.hist([movies['release_year'], shows['release_year']],
         bins=30,
         color=['lightgreen', 'lavender'],
         edgecolor = 'pink',
         alpha=0.8, 
         label =['Movies', 'TV-Shows']
        )
plt.title('Netflix release by Year and Type')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles Released')
plt.grid(axis='y', color='lightblue', linestyle='--', linewidth=0.8)
plt.tight_layout()
plt.legend(title="Movies/shows")
plt.savefig("netflix_showvsmovies.jpg", dpi=300, bbox_inches='tight')
plt.show()
```

# Use netflix_data.csv dataset 
    ## print number of Movies and TV Shows available on Netflix 
    ## Display movies and TV shows proportion in pie plot 
    ## Save file as pdf

```python
import pandas as pd
import matplotlib.pyplot as plt 

df= pd.read_csv('netflix_data.csv')
df = df.dropna(subset='type')

movies=df[df['type']=='Movie']
movies_count= (movies['show_id'].count())
shows=df[df['type']=='TV Show']
shows_count = (shows['show_id'].count())
print (movies_count) 
print (shows_count)
```

```python
type_counts=df['type'].value_counts()
print (type_counts)
```

```python
plt.figure(figsize=(7,7))
plt.pie(type_counts, labels=type_counts.index,
        colors=['pink', 'lavender'], autopct='%1.1f%%', 
        startangle=140, 
        explode=(0.05, 0.05)
       )
plt.title("Movies and TV Shows Distribution") 
plt.legend()
plt.savefig('netflix_pieplot.pdf',dpi=300)
plt.show()
```
