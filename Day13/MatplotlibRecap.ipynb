```python
import matplotlib.pyplot as plt 
```

```python
months=['Jan', 'Feb', 'Mar', 'April']
sales_value=[2000,3000,2500,3600]
```

# plt.plot(x,y,color=<>,linestyle=<>, linewidth=<>, marker=<>, label=<>)

```python
plt.plot(months,sales_value, color='pink', linestyle='--', linewidth = 2, marker='o', label="monthly sales data")
plt.title('Sales Data')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.xlim(0,4)
plt.ylim(2000,5000)
plt.grid(True)
plt.legend(loc='lower right')
plt.show()
```

```python
import numpy as np 
```

```python
x=np.linspace(0,2,100) #Generate 100 evenly spaced lines between 0 and 2
plt.plot(x,x,label='Linear') #second x is y, first x is above line
plt.plot(x,x**2,label='Quadratic')
plt.plot(x,x**3,label='Cubic')
plt.xlabel('X Lines')
plt.ylabel('Y Lines')
plt.title('Simple Lines Plot') 
plt.legend()
plt.show()
```

# Bar Plot Example 
    ## plt.bar(x,y,color=<>,edgecolor=<>)


```python
categories = ['Electronics', 'Mobiles', 'Books', 'Groceries', 'Home Decoration']
sales = [5000,60000,30000,90000,4000]
bar_colors = ('pink', 'lightgreen', 'white', 'skyblue', 'lavender')
plt.bar(categories,sales,color=bar_colors, edgecolor='red')
plt.xlabel('categories')
plt.ylabel('sales')
plt.title('Categories vs Sales') 
plt.grid(axis='y', linestyle='--', alpha=0.2, color='purple')
plt.tight_layout()
#plt.legend(loc='upper left')
plt.show()
```

```python
categories = ['Electronics', 'Mobiles', 'Books', 'Groceries', 'Home Decoration']
sales = [5000,60000,30000,90000,4000]
bar_colors = ('pink', 'lightgreen', 'white', 'skyblue', 'lavender')
plt.bar(categories,sales,color=bar_colors, edgecolor='red')
plt.xlabel('categories')
plt.ylabel('sales')
plt.title('Categories vs Sales') 
plt.grid(axis='y', linestyle='--', alpha=0.2, color='purple')
plt.xticks(rotation=30, fontsize=8, color='magenta')
plt.yticks([20000,40000,60000,80000,100000],['20K','40K','60K','80K','100K'],fontsize=8, color='orange')
plt.tight_layout()
#plt.legend(loc='upper left')
plt.show()
```

# Scatter Plot 
    ## plt.scatter(x,y,s=<size>,c=<color>,marker=<symbol "x", "o">,alpha=<transparency>,edgecolor,label)

```python
height_cm=[150,155,160,165,170,175,180,185]
weight_kg=[50,53,56,59,62,66,68,72]
plt.figure(figsize=(8,5))
plt.scatter(height_cm, weight_kg, s=100, c='pink', marker='o', alpha=0.75, label = "weight vs height") 
plt.title("Scatter plot of weight vs height") 
plt.xlabel("Height") 
plt.ylabel("Weight")
plt.xticks(rotation=0, fontsize=12, color='orange')
plt.yticks([50,55,60,65,70,75], ['50kg','55kg','60kg','65kg','70kg','75kg'], fontsize=12, color='lime')
plt.legend(loc="best")
plt.grid(color='lightblue', alpha=0.5, linestyle='dotted')
plt.show()
```
