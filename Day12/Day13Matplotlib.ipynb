# Scattered Plot Example - using random integer

```python
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
temp=np.random.randint(20,100,size=50)
plt.scatter(x,y,c=temp,cmap='viridis',s=120, edgecolor='black') #cmap='plasma' or 'viridis' - '<>' is color theme
plt.colorbar(label='Temperature')
plt.title("Scatter plot colored by temperature")
plt.xlabel('X-random')
plt.ylabel('Y-random')
plt.show()
```

```python
height_cm=[150,155,160,165,170,175,180,185]
weight_kg=[50,53,56,59,62,66,68,72]
age_year=[20,22,25,28,30,35,38,40]
```

```python
#height_cm=[150,155,160,165,170,175,180,185]
#weight_kg=[50,53,56,59,62,66,68,72]
#age_year=[20,22,25,28,30,35,38,40]

plt.figure(figsize=(8,5))
scatter=plt.scatter(height_cm, weight_kg, c=age_year, cmap='rainbow', s=120, alpha=0.9)
plt.colorbar(scatter, label='Age (years)')
plt.title('Scatter plot: Height vs Weight (colored by age)')
plt.xlabel('Height(cm)')
plt.ylabel('Weight(kg)')
plt.grid(color='blue', linestyle='--', linewidth=0.7)
plt.show()
```

# Pie Plot Example 
    ## plt.pie(x,labels='',colors='',autopct='',startangle='',shadow=,explode=,radius)
    ## autopct - decimal places -- '%1.0f%%' meaning no decimal places, 1.1 is one decimal place

```python
depts=['HR', 'Web Development', 'App Development', 'Agile-Scrum', 'Testing']
expenses=[12000,25000,18000,15000,20000]
```

```python
plt.figure(figsize=(7,7))
```

```python
plt.pie(expenses,labels=depts, autopct='%1.1f%%', startangle=140, explode=(0, 0.05, 0, 0, 0), shadow=True)
plt.title("Department Expenses Report")
plt.show()
```

# Add colormap

```python
plt.pie(expenses,labels=depts, colors=plt.cm.rainbow([0.1,0.3,0.5,0.7,0.9]), autopct='%1.1f%%', startangle=140, explode=(0, 0.05, 0, 0, 0), shadow=True)
plt.title("Department Expenses Report")
plt.show()
```

# Create a pie plot using following data: 
## sales_data={'Mobile':45000, 'Clothing':30000, 'Groceries':35000, 'Home decor':20000}
### show percentage with two decimal 
### use colormap to give different color to your pie



```python
product_sales={'Mobile':45000, 'Clothing':30000, 'Groceries':35000, 'Home decor':20000}
products=list(product_sales.keys())
sales=list(product_sales.values())
plt.figure(figsize=(7,7))

plt.pie(sales,labels=products, colors=plt.cm.rainbow([0.1,0.25,0.37,0.8]), 
        autopct='%1.2f%%', startangle=140, explode=(0, 0, 0, 0), shadow=True)
plt.legend(loc='best')
plt.title("Product Sales Report")
plt.show()
```

# Save file

```python
product_sales={'Mobile':45000, 'Clothing':30000, 'Groceries':35000, 'Home decor':20000}
products=list(product_sales.keys())
sales=list(product_sales.values())
plt.figure(figsize=(7,7))

plt.pie(sales,labels=products, colors=plt.cm.rainbow([0.1,0.25,0.37,0.8]), 
        autopct='%1.2f%%', startangle=140, explode=(0, 0, 0, 0), shadow=True)
plt.legend(loc='best')
plt.title("Product Sales Report")
plt.savefig('categorieswisesale.png', dpi=300, bbox_inches='tight')
plt.show()
```

# Histogram

```python
marks=[10,30,40,50,60,70,61,89,43,59,98,67,75,90,56,68,49,52,75,87,90,11,100]
print(len(marks))
#plt.style.use('ggplot')
plt.figure(figsize=(8,5))
#bar_colors=['lavender','pink','lime', 'skyblue', 'white']
plt.hist(marks,bins=5, color="pink", edgecolor='red', alpha=0.8)
plt.title("Histogram for Student Marks")
plt.xlabel("Marks")
plt.ylabel("Number of students")
plt.grid(color='blue',linestyle='--', linewidth=0.7)
plt.show()
```
