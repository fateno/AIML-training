# First example to cerate line chart from out data 
    ## Students.xlsx for fetching the data 
    ## Generate line chart based on student name and student marks

```python
import pandas as pd
```

```python
df = pd.read_excel("students1.xlsx")
```

```python
df
```

```python
names = df['Name']
marks = df['Marks']
```

```python
plt.plot(names,marks,marker='s', color='green')
plt.title('Student Marks Line Chart')
plt.xlabel('Student Name')
plt.ylabel('Student Marks')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.plot(names,marks,marker='v', color='orange', linestyle='-.')
plt.title('Student Marks Line Chart')
plt.xlabel('Student Name')
plt.ylabel('Student Marks')
plt.grid(True)
plt.show()
```
