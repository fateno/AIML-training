```python
players = ['Faten', 'Sofea', 'Bluey', 'Bingo', 'Chili']
match1_score=[45,67,78,89,10]
match2_score=[55,78,40,80,97]
match3_score=[30,50,70,85,90]
plt.figure(figsize=(8,5))
plt.scatter(players,match1_score,color='pink', label='Match One', s=100)
plt.scatter(players,match2_score,color='limegreen', label='Match Two', s=100, marker='x')
plt.scatter(players,match3_score,color='lavender', label='Match Three', s=100, marker='p')
plt.xlabel('Players')
plt.ylabel('Score')
plt.legend(loc='best')
plt.tight_layout()
plt.grid(True, linestyle='dotted',alpha=1)
plt.show()
```

# Write a comparison that shows average calories burned for each workout type

```python
workout_type=df['Workout_Type']
avg_calories=df.groupby('Workout_Type')['Calories_Burned'].mean()
plt.figure(figsize=(8,5))
plt.scatter(avg_calories.index, avg_calories.values, label='calories burned', color='pink', s=100, marker='p')
plt.title('Comparison workout vs calories burned') 
plt.xlabel('Workout Type')
plt.ylabel('Average Calories Burned')
plt.legend(loc='best')
plt.tight_layout()
plt.grid(True, linestyle='dotted',alpha=1)
plt.show()
```
