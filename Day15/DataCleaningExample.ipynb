```python
import pandas as pd 
```

```python
data={'Students':['Amit', 'Anam', 'Chi', 'Sam', 'Ravi', 'Jai'],
'Marks':[85, None, 65,75, None, 55]
}
```

```python
df=pd.DataFrame(data)
df
```

```python
print ('Missing Data')
print ("Number of missing values are:" ,df.isnull().sum())
```

```python
average_marks=df['Marks'].mean()
print (average_marks)
```

```python
df['Marks'].fillna(average_marks, inplace=True)
#df['Marks'].fillna("0.0", inplace=True)
```

```python
print ('After fillna() call: ')
df
```

# How we will use in our projects - fillna() example: 

```python
data={'Students':['Amit', 'Anam', 'Chi', 'Sam', 'Ravi', 'Jai'],
'Marks':[85, None, 65,75, None, 55]
}
df=pd.DataFrame(data)
print ('Original data:')
df 
```

```python
df.fillna(df['Marks'].mean(), inplace=True)
print ('After fillna.() data:')
df 
```

```python
data={'Students':['Amit', 'Anam', 'Chi', 'Sam', 'Ravi', 'Jai'],
'Marks':[85, None, 65,75, None, 55]
}
df=pd.DataFrame(data)
print ('Original data:')
df 
```
