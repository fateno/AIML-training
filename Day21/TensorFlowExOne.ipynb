**Matrix Multiplication**
We will use tf.matmul() method

**Rules:**
1. Inner dimensions must match
(m,n)@(n,p)
for e.g.
(5,7)@(8,9) it will not work because n is different
(5,8)@(8,9) it will work because inner dimension is the same


2. The result matrix has the shape of outer dimension meaning A(m,n) x B(n,p) = (m,p) (4,7)x(7,5)=(4,5)
